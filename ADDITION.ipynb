```python
import numpy as np
import pandas as pd
import time

a = input('Enter range:').split()
a = list(map(int,a))
performance = pd.DataFrame(columns=['Numbers','Answer','Entered Answer','Time Taken'])


size = int(input('Enter Size:'))
begin_time = time.time()
It = int(input('Iterations:'))

print('----------------------------------------------------------------------')
print('----------------------------------------------------------------------')



for i in range(It):
    
    start_time = time.time()
    x = np.random.randint(a[0],a[1],size = size)
    print(i,')',x)

    ans = sum(x)

    entered = int(input('Enter the sum:'))

    
    if entered == ans:
        flag = 1
    else:
        flag = 0
        
        
    if flag == 1:
        print('Correct!')
        end_time = time.time()
        df= {'Numbers':x,'Answer':sum(x),'Entered Answer':ans,'Time Taken':(end_time-start_time)}
        performance = performance.append(df, ignore_index=True)
        print('Time Taken:',end_time-start_time,'secs') 
        print('----------------------------------------------------------------------')
        continue
        print('END')
        
        
    else:
        end_time = time.time()
        df= {'Numbers':x,'Answer':sum(x),'Entered Answer':ans,'Time Taken':(end_time-start_time)}
        performance = performance.append(df, ignore_index=True)
        print('over')
        print('Correct Ans is :',sum(x))
        break
        

end_ti = time.time()
print('Total time:',(end_ti-begin_time)/60,'mins')

import openpyxl
dt = list(time.localtime())
dt
da=str(dt[:3])
ti=str(dt[3:6])
dati = da+ti
filename = 'Addition'+dati+'.xlsx'


performance.to_excel(filename, index = False)
```
